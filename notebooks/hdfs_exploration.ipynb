```python
from collections import Counter
```

```python
def get_token_counts(series):
    counter = Counter()
    for seq in series:
        tokens = seq.split()
        counter.update(tokens)
    return counter
    
```

```python
normal_counts = get_token_counts(df_normal["event_sequence"])
abnormal_counts = get_token_counts(df_abnormal["event_sequence"])
normal_counts
```

```python
abnormal_counts
```

```python
for token in abnormal_counts:
    if abnormal_counts[token] > normal_counts.get(token, 0) * 1.5:
        print(token)
```

```python
total_normal_count = sum(normal_counts.values())
total_normal_count
```

```python
total_abnormal_count = sum(abnormal_counts.values())
total_abnormal_count
```

```python
normal_freq = pd.Series(normal_counts) / total_normal_count
normal_freq
```

```python
abnormal_freq = pd.Series(abnormal_counts) / total_abnormal_count
abnormal_freq
```

```python
ratio = abnormal_freq / normal_freq
ratio.sort_values(ascending = False).head(10)
```
